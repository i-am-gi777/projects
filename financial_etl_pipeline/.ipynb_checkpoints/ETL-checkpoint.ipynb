```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# ==================================================
# Spark Session (Hive + LocalStack)
# ==================================================
spark = (
    SparkSession.builder
    .appName("FinancialLedgerConsolidation")

    # Hive Support (supports your first process)
    .config("spark.sql.catalogImplementation", "hive")
    .enableHiveSupport()

    # LocalStack S3 Configuration
    .config(
        "spark.jars.packages",
        "org.apache.hadoop:hadoop-aws:3.3.4,"
        "com.amazonaws:aws-java-sdk-bundle:1.12.262"
    )
    .config("spark.hadoop.fs.s3a.endpoint", "http://localhost:4566")
    .config("spark.hadoop.fs.s3a.access.key", "test")
    .config("spark.hadoop.fs.s3a.secret.key", "test")
    .config("spark.hadoop.fs.s3a.path.style.access", "true")
    .config("spark.hadoop.fs.s3a.connection.ssl.enabled", "false")

    .getOrCreate()
)

# ==================================================
# Create Database (Glue Catalog Simulation)
# ==================================================
spark.sql("CREATE DATABASE IF NOT EXISTS financial_db")

# ==================================================
# ✅ Your S3 Paths (Replaced)
# ==================================================
S3_BUCKET = "financial-pipeline-bucket"

SALES_PATH = "s3a://financial-pipeline-bucket/data/sales.csv"
TAX_PATH = "s3a://financial-pipeline-bucket/data/tax.csv"
OUTPUT_PATH = "s3a://financial-pipeline-bucket/financial-data/"

# ==================================================
# Load Sales Data
# ==================================================
print("Loading Sales Data...")

sales_df = (
    spark.read.option("header", "true")
    .csv(SALES_PATH)
    .withColumn("sale_amount", col("sale_amount").cast("double"))
    .withColumn("sale_date", to_date(col("sale_date"), "yyyy-MM-dd"))
)

# ==================================================
# Load Tax Data
# ==================================================
print("Loading Tax Data...")

tax_df = (
    spark.read.option("header", "true")
    .csv(TAX_PATH)
    .withColumn("tax_amount", col("tax_amount").cast("double"))
    .withColumn("tax_rate", col("tax_rate").cast("double"))
)

# ==================================================
# Business Logic
# ==================================================
print("Joining Data...")

consolidated_df = (
    sales_df.join(tax_df, "transaction_id", "left")
    .withColumn(
        "total_amount",
        round(
            coalesce(col("sale_amount"), lit(0)) +
            coalesce(col("tax_amount"), lit(0)), 2
        )
    )
    .withColumn(
        "net_profit_margin",
        round(
            (col("sale_amount") -
             coalesce(col("tax_amount"), lit(0)))
            / col("sale_amount") * 100, 2
        )
    )
    .withColumn("ledger_date", current_date())
)

consolidated_df.show()

# ==================================================
# Write to S3 (Data Lake)
# ==================================================
print("Writing Parquet to S3...")

consolidated_df.write \
    .mode("overwrite") \
    .parquet(OUTPUT_PATH)

# ==================================================
# Register Hive Table (Glue Style)
# ==================================================
print("Creating Hive Table...")

consolidated_df.write \
    .mode("overwrite") \
    .format("parquet") \
    .saveAsTable("financial_db.financial_ledger")

spark.sql("SHOW TABLES IN financial_db").show()

print("ETL Pipeline Complete!")

spark.stop()
```
